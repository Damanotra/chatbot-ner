```python
import pandas as pd
```

```python
bilstm_cnns_df = pd.read_csv('bilstm_cnns_w2v.csv', index_col=[0]).reset_index().loc[:20]
bilstm_df = pd.read_csv('bilstm_w2v.csv', index_col=[0]).reset_index().loc[:20]
```

```python
bilstm_df
```

```python
bilstm_df.columns = ['label', 'p', 'r', 'f', 's']
bilstm_cnns_df.columns = ['label', 'p', 'r', 'f', 's']
```

```python
bilstm_df['pos'] = bilstm_df['label'].map(lambda x: x.split('-')[0])
bilstm_df['ne'] = bilstm_df['label'].map(lambda x: x.split('-')[-1])

bilstm_cnns_df['pos'] = bilstm_cnns_df['label'].map(lambda x: x.split('-')[0])
bilstm_cnns_df['ne'] = bilstm_cnns_df['label'].map(lambda x: x.split('-')[-1])
```

```python
NE_bilstm_df = bilstm_df[['pos', 'f']]
NE_bilstm_cnns_df = bilstm_cnns_df[['pos', 'f']]
```

```python
NE_bilstm_df = NE_bilstm_df.groupby(['pos']).mean().reset_index()
NE_bilstm_cnns_df = NE_bilstm_cnns_df.groupby(['pos']).mean().reset_index()
```

```python
NE_bilstm_cnns_df
```

```python
df = pd.merge(NE_bilstm_df, NE_bilstm_cnns_df, on='pos', how='inner')
```
